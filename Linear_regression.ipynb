This is a demo to do the bootstrapped linear regression using the tool
The raw data should be stored inside a specific folder

```python
from stat_analysis import Regression
import numpy as np
```

```python
work_path = r'./data/AS87/*'
# The folder containing the raw data can be defined here
```

```python
param_name_list = ['Rt', 'Rp', 'Rv', 'Rzjis', 'Ra', 'Rq', 'Rsk', 'Rku', 'delta_ANAM']

# the roughness parameter we would like to check can be defined here, and those name should
# match up with the functions in afm_func_1d.py
```

```python
hyperparam_for_roughness = {  # the hyper_parameter for the calcuation of roughness
    'tau_list': np.arange(1, 10),  # for the calculation of delta_ANAM
    'alpha': 2,  # for the calculation of delta_ANAM
    'window_length': 31, # for the smoothing of line profile if needed
    'polyorder': 5,  # for the smoothing of line profile if needed
    'size': 256,  # for the generation of line profile
    'stride': 32,  # for the generation of line profile
    'window': 4,  # for the generation of line profile
    'use_smooth': False  # for the smoothing of line profile if needed
}
```

```python
AS87 = Regression(work_path, hyperparam_for_roughness, param_name_list)
s_values, r_values, X, Y, s_ci_pi, r_ci_pi = AS87.main()
AS87.data_visualization() # The plot of data will be saved into './figures/data'
AS87.res_visualization() # The plot of results will be saved into './figures/res'
```
